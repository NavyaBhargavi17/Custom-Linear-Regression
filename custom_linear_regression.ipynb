```python
# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2: Custom Linear Regression class using Least Squares Method
class LinearRegressionCustom:
    def __init__(self):
        self.slope = 0
        self.intercept = 0

    def fit(self, X, y):
        # Calculate means
        x_mean = np.mean(X)
        y_mean = np.mean(y)

        # Calculate slope (m) and intercept (c)
        numerator = sum((X[i] - x_mean) * (y[i] - y_mean) for i in range(len(X)))
        denominator = sum((X[i] - x_mean) ** 2 for i in range(len(X)))
        self.slope = numerator / denominator
        self.intercept = y_mean - self.slope * x_mean

    def predict(self, X):
        # Predict y = m*x + c for each value in X
        return [self.slope * x + self.intercept for x in X]

    def get_params(self):
        # Return slope and intercept
        return self.slope, self.intercept
```

```python
# Step 3: Sample data (X and y)
X = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
```

```python
# Step 4: Train the custom linear regression model
model = LinearRegressionCustom()
model.fit(X, y)

# Get slope and intercept
slope, intercept = model.get_params()
print(f"Slope (m): {slope:.2f}")
print(f"Intercept (c): {intercept:.2f}")
```

```python
# Step 5: Predict values for X
predictions = model.predict(X)
print("Predicted values:", predictions)
```

```python
# Step 6: Plot original data points and regression line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, predictions, color='red', label='Regression Line')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Custom Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
